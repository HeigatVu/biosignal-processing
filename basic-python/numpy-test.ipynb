# Numpy array

```python
np_vector = np.array([1, 2, 3])
print(np_vector)
print(type(np_vector))
print(np_vector.shape)
```

```python
np_matrix = np.array([
    [1, 2, 3,],
    [8, 3, 1],
])

print(np_matrix)
print(np_matrix.shape)
```

```python
np_vector = np.zeros((4, ))
print(np_vector)
print(np_vector.shape)
```

```python
unit_matrix = np.eye((4))
print(unit_matrix)
```

```python
np_vector = np.arange(1, 22, 3)
print(np_vector)
```

# Indexing and slicing

```python
np_matrix = np.array([
    [1, 2, 3],
    [9, 7, 0]
])

print(np_matrix)
print(np_matrix.shape)

print(np_matrix[0, :2])
print(np_matrix[:, 0])
print(np_matrix[-1])
print(np_matrix[-1, :2])
```

```python
print(np_matrix[:1, :2])
print(np_matrix[:1, :2].shape)
```

```python
np_matrix = np.random.random((4, 6))
print(np_matrix.shape)
print(np_matrix)
```

```python
print(np_matrix[0:1].shape)
print(np_matrix[0:1])
```

```python
print(np_matrix[0:1][-1].shape)
print(np_matrix[0:1][-1])
```

```python
print(np_matrix[0:1][-1][1:3])
```

# Broadcasting

```python
np_vector = np.array([1, 2, 3])
np_vector = np_vector + 3
print(np_vector)
```

```python
np_matrix = np.array([
    [4, 5, 6],
    [7, 8, 9],
])

print(np_matrix*2)
```

```python
np_vector_range = np.arange(0, 20, 3)
print(np_vector_range)
print(np_vector_range*3)
```

```python
a = np.ones((3, 5))
print(f"a: {a}")
b = np.arange(5)
print(f"b: {b}")
print(a + b)
```

```python
a_new = np.arange(5)
print(a_new.shape)
a_new = a_new.reshape(5, 1)
print(a_new.shape)
print(a_new)

b_new = np.arange(5)
print(b_new.shape)
print(b_new)

c_new = a_new + b_new
print(c_new.shape)
print(c_new)
```

# Iterator

```python
np_vector = np.random.random(5, )
print(np_vector)
for x in np.nditer(np_vector):
    print(x)
```

```python
np_matrix = np.random.random((2, 3))
print(np_matrix)
for x in np.nditer(np_matrix):
    print(x)
```

```python
np_cube = np.random.random((2, 2, 2))
print(np_cube)
for x in np.nditer(np_cube):
    print(x)
```

```python
np_matrix = np.random.random((2, 3))
print(np_matrix)
for x in np.nditer(np_matrix, flags=["external_loop"], order='F'):
    print(x)
```

```python
for x in np.nditer(np_matrix, flags=["external_loop"], order='C'):
    print(x)
```

# Change size of array

```python
a = np.array([
    [1, 2, 3, 4],
    [5, 6, 7, 8],
])
print(a.shape)
```

```python
b = a.reshape(4, 2)
print(b)
```

```python
np_cube = np.random.random((2, 2, 2))
print(np_cube)
print(np_cube.shape)
```

```python
tmp = np_cube.flatten()
print(tmp)
print(tmp.shape)
```

```python
a_test = np.array([
    [1, 2],
    [4, 5],
])
print(a_test.shape)
b_test = np.expand_dims(a_test, axis=0)
print(b_test.shape)
print(b_test)
```

```python
np_cube = np.random.random((1, 2, 2, 1))
print(np_cube)
result = np.squeeze(np_cube, axis=0)
print(result)
print(result.shape)
```

# Transpose

```python
np_matrix = np.array([
    [1, 2, 3, 4],
    [5, 6, 7, 8],
])
print(np_matrix.shape)
print(np_matrix.T)
print(np_matrix.T.shape)
```

# Products

```python
a = np.array([
    [1, 2, 3],
    [3, 1, 0],
])

b = np.array([
    [2, 1, 3],
    [1, 3, 4],
    [3, 1, 0],
])

print(a.dot(b))
```

```python
a = np.array([1, 2, 3])

b = np.array([3, 4, 5])

print(np.inner(a, b))
```

```python
A = np.array([
    [2, 0, 6],
    [1, -1, 2],
])
B = np.array([
    [8, -3, 2],
    [4, 1, -5]
])

print(np.tensordot(A, B))
```

# Norms

```python
A = np.array([1, 2, 3])
print(np.linalg.norm(A))
```

```python
print(np.linalg.norm(A, ord=1))
```
